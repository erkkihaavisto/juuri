```python
# Tuodaan Pandas ja Matplotlib -ohjelmakirjastot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from matplotlib.ticker import PercentFormatter

# Tuodaan prosenttimuotoilija, jota voidaan käyttää kun muotoillaan kuvioiden prosenttiakseleita
from matplotlib.ticker import PercentFormatter

# Prosenttimuotoilua - Jos prosentit desimaalilukuja
myFmt1 = PercentFormatter(xmax=1, decimals=0, symbol=' %')

# Prosenttimuotoilua - Jos prosentit on jo valmiiksi kerrottu sadalla
myFmt100 = PercentFormatter(xmax=100, decimals=0, symbol=' %')

# Avataan data - Rahastoraporttidata 2019-07
df = pd.read_excel('http://taanila.fi/Rahastoraporttidata_201907.xlsx')

# Tulostetaan datan 5 ensimmäistä riviä, jos mitään ei tulostu niin yläpuolella on jo sattunut virhe
df.head()
```

```python
# Dataa on 1350 riviä ja 39 saraketta. Melko paljon!
# Tehdään df:ään uusi kolumni: 'Pääomaluokka', joka on vain 'RR2009_class' -kolumni suomennettuna

# Tässä on hyödynnetty lähdetiedoston selitteitä ja Excelin CONCAT-funktiota hyväksi, 
# jotta saadaan nopeasti monta riviä oikeaan muotoon.

df['Pääomaluokka'] = df['RR2009_class'].replace({'AA_EUROPE' : 'Yhdistelmärahastot Eurooppa',
                                                 'AA_GLOBAL' : 'Yhdistelmärahastot Maailma',
                                                 'ALT_HEDGE' : 'Vaihtoehtoiset Hedge',
                                                 'ALT_OTHER' : 'Vaihtoehtoiset Muut',
                                                 'BOND_EMER' : 'Pitkä Korko Kehittyvät Markkinat',
                                                 'BOND_GOVT' : 'Pitkä Korko Valtioriski Euroalue',
                                                 'BOND_GOVT_GLOBAL' : 'Pitkä Korko Valtioriski Maailma',
                                                 'BOND_HIGH_EURO' : 'Pitkä Korko Korkeariskiset Yrityslainat Eurooppa',
                                                 'BOND_HIGH_GLOBAL' : 'Pitkä Korko Korkeariskiset Yrityslainat Maailma',
                                                 'BOND_IG_EURO' : 'Pitkä Korko Luokitellut Yrityslainat Eurooppa',
                                                 'BOND_IG_GLOBAL' : 'Pitkä Korko Luokitellut Yrityslainat Maailma',
                                                 'EQ_EMER_EUROPE' : 'Osakerahastot Keski- ja Itä-Eurooppa',
                                                 'EQ_EMER_FAREAST' : 'Osakerahastot Kehittyvät Markkinat Kaukoitä',
                                                 'EQ_EMER_GLOBAL' : 'Osakerahastot Kehittyvät Maailma',
                                                 'EQ_EMER_LATAM' : 'Osakerahastot Latinalainen Amerikka',
                                                 'EQ_EMER_SAHARA' : 'Osakerahastot Lähi-Itä & Afrikka',
                                                 'EQ_EUROPE' : 'Osakerahastot Eurooppa',
                                                 'EQ_FINLAND' : 'Osakerahastot Suomi',
                                                 'EQ_GLOBAL' : 'Osakerahastot Maailma',
                                                 'EQ_JAPAN' : 'Osakerahastot Japani',
                                                 'EQ_NORAM' : 'Osakerahastot Pohjois-Amerikka',
                                                 'EQ_NORDIC' : 'Osakerahastot Pohjoismaat',
                                                 'EQ_PAC' : 'Osakerahastot Tyynenmeren Alue',
                                                 'EQ_SECTOR_CONSUMER' : 'Osakerahastot Sektorit Kulutushyödykkeet ja Päivittäistavarat',
                                                 'EQ_SECTOR_ENERGY' : 'Osakerahastot Sektorit Energia ja Yleishyödylliset Palvelut',
                                                 'EQ_SECTOR_FINANCIAL' : 'Osakerahastot Sektorit Rahoitus ja Kiinteistöt',
                                                 'EQ_SECTOR_HEALTH' : 'Osakerahastot Sektorit Terveydenhuolto',
                                                 'EQ_SECTOR_INDUSTRIAL' : 'Osakerahastot Sektorit Teollisuus ja Perusteollisuus',
                                                 'EQ_SECTOR_TECHNO' : 'Osakerahastot Sektorit Teknologia & Telekommunikaatio',
                                                 'MM_CASH' : 'Käteisrahastot',
                                                 'MM_MM' : 'Rahamarkkinarahastot',
                                                 'MM_OTHER' : 'Lyhyen Koron Rahastot Muut',})

df.head()
```

```python
# Järjestetään rahastoyhtiöt rahastopääoman mukaan pienimmästä suurimpaan.

df1 = df.groupby('company')['fund_size'].sum().sort_values(ascending=False).round(2)

df1
```

```python
# Edellinen toimii, mutta haluammekin vain Suomalaiset rahastoyhtiöt.
# Tehdään dataframe, johon otamme mukaan kaikki rivit joilla on COUNTRY_FI sarakkeessa 'Suomi'
df1 = df[df['COUNTRY_FI'] == 'Suomi']


# Poistan ylimääräiset osuussarjat käyttäen lähdedatan top_id luokittelua.
# Mikäli näitä ei poista, näkyisi joidenkin rahastojen pääoma tuplana tai jopa triplana.

df2 = df1.drop_duplicates(subset ="top_id", 
                     keep = 'first', inplace = False)


df3 = df2.groupby('company')[['fund_size']].sum().astype(int).sort_values('fund_size',ascending = False)

df3.columns.name = ''

df3
```

```python
# Lisätään taulukkoon prosenttisarake
df3['%'] = df3/df3.sum()*100

# Muotoillaan prosentit kahden desimaalin tarkkuuteen
df3.style.format({'%': '{:.2f} %'})

```

```python
# Tehdään pivot-taulukko, ja katsotaan kuinka rahastopääomat allokoituu eri pääomaluokkien välillä yhtiötasolla.
# Indeksiin asetetaan rahastoyhtiö, ja kolumneihin pääomaluokka
df4 = df[df['COUNTRY_FI'] == 'Suomi']
df4 = df4.pivot_table(values='fund_size', index='company', columns='Pääomaluokka', aggfunc=sum)

df4.style.format('{:.0f}')
```

```python
# Taulukossa oli paljon nan-arvoja, koitetaan tehdä listaus eri tavalla niin, että kumpikin muuttuja on indeksissä
# Tehdään pivot-taulukko, ja katsotaan kuinka rahastopääomat allokoituu eri pääomaluokkien välillä yhtiötasolla.
df4 = df[df['COUNTRY_FI'] == 'Suomi']
df4 = df4.pivot_table(values='fund_size', index=['company', 'Pääomaluokka'], aggfunc=sum)

df4.style.format('{:.0f}')
```
